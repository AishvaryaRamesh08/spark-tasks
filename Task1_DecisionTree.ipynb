## **Prediction using Decision Tree Algorithm**
Aim: To build a Decision Tree model which can predict right class when new data is fed

**Importing the necessary libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

**Importing the data**

```python
data = pd.read_csv('Iris.csv')
data
```

```python
data.info()
```

```python
data.drop(columns='Id', inplace=True)
data
```

```python
data.Species.value_counts()
```

```python
sns.set_style('whitegrid')
sns.scatterplot(x=data['PetalLengthCm'], y=data['PetalWidthCm'], hue=data['Species'])
```

```python
sns.scatterplot(x=data['SepalLengthCm'], y=data['SepalWidthCm'], hue=data['Species'])
```

```python
sns.pairplot(data, hue='Species')
```

**Data preparation**

```python
iris_class = {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}
data['Species'] = [iris_class[i] for i in data['Species']]
data
```

```python
X = data.iloc[:, 0:-1]
X.head()
```

```python
y = data.iloc[:, -1]
y
```

**Defining and fitting the model** 

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

**Visualizing the model**

```python
fn = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
cn = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(model,
               feature_names = fn, 
               class_names=cn,
               filled = True);
fig.savefig('imagename.png')
```
